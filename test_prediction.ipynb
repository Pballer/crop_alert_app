```python
def make_prediction(file_path, model):
    image = cv2.imread(file_path)
    image = cv2.resize(image, (224,224))
    image = image * 1.0
    image = np.reshape(image,[1, 224, 224, 3])
    prediction = model.predict_classes(image, batch_size=1)
    print(model.predict(image))
    return prediction, image
```

```python
def tf_image_load(file_path, model):
    image = load_img(file_path, target_size=(224, 224))
    x = img_to_array(image)
    x = np.expand_dims(x, axis=0)
    images = np.vstack([x])
    pred_classes = model.predict(images, batch_size=10)
    classes = pred_classes.argmax(axis=-1)
    return classes[0]
```
